```python
import pandas as pd

# Load the merged dataset
df = pd.read_csv("merged_output.csv")

# Show basic info and first few rows
print(df.info())
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.histplot(df['ac_output'], kde=True)
plt.title("Distribution of AC Output")
plt.show()

sns.histplot(df['dc_output'], kde=True)
plt.title("Distribution of DC Output")
plt.show()


sns.histplot(df['solar_radiation_GHI'], kde=True)
plt.title("Distribution of Solar Radiation (GHI)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


features = [
    'temperature_max', 'temperature_min', 'humidity_mean', 'cloud_cover_mean',
    'windspeed_max', 'windgusts_max', 'precipitation_total',
    'precipitation_hours', 'sunshine_hours', 'solar_radiation_GHI',
    'daylight_hours', 'ac_output', 'dc_output'
]


corr_matrix = df[features].corr()


plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import folium

# Read your merged data
df = pd.read_csv("merged_output.csv")

# Create a base map centered on the USA
m = folium.Map(location=[39.5, -98.35], zoom_start=4)  # No attribution needed

# Add all lat/lon points to the map
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=2,
        color="blue",
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

# Save and display map
m.save("us_locations_map.html")
m
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("merged_output.csv")

# Group by month and calculate average AC output
monthly_avg_ac = df.groupby("month")["ac_output"].mean()

# Define month labels
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Plot the average AC output by month as a line graph
plt.figure(figsize=(10, 6))
monthly_avg_ac.plot(kind='line', marker='o', color='skyblue')
plt.title("Average Monthly AC Output (kWh)")
plt.xlabel("Month")
plt.ylabel("AC Output (kWh)")
plt.xticks(ticks=range(1, 13), labels=month_labels)
plt.grid(True)
plt.tight_layout()
plt.show()
```
